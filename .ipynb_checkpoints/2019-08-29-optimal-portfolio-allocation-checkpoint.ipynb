Import required python libraries and add some formatting options

```python
from math import sqrt

import numpy as np
import pandas as pd
from scipy.optimize import minimize
import matplotlib.pyplot as plt

float_formatter = lambda x: "%.4f" % x
np.set_printoptions(formatter={'float_kind':float_formatter})
```

Define a function that calculates optimal portfolio by minimising the volatility for a given level of risk

```python
def optimise_portfolio(target_mu=0.045, corr_factor=1):
    """ Returns a string decribing optimial portfolio allocation
    
    target_mu: float, target portfolio return
    corr_factor: float, value to scale correlation up by, useful for q1.3
    
    """
    
    mu = np.matrix([
        [0.02], 
        [0.07], 
        [0.15], 
        [0.20],
    ])

    sigma = np.diag([
        0.05, 
        0.12, 
        0.17, 
        0.25,
    ])

    correlation = np.matrix([
        [1.00, corr_factor * 0.30, corr_factor * 0.30, corr_factor * 0.30],
        [corr_factor * 0.30, 1.00, corr_factor * 0.60, corr_factor * 0.60],
        [corr_factor * 0.30, corr_factor * 0.60, 1.00, corr_factor * 0.60],
        [corr_factor * 0.30, corr_factor * 0.60, corr_factor * 0.60, 1.00],
    ])

    covariance = sigma @ correlation @ sigma

    start_weights = np.matrix([
        [0.25], 
        [0.25], 
        [0.25], 
        [0.25],
    ])

    def calc_sigma(my_weights):
        return sqrt(my_weights.T @ covariance @ my_weights)

    r = minimize(calc_sigma, start_weights, tol=0.000000001,
             constraints=(
                 {
                     'type': 'eq', 
                     'fun': lambda weights: 1.0 - np.sum(weights)
                 },
                 {
                     'type': 'eq', 
                     'fun': lambda weights: target_mu - (weights.T @ mu).item((0, 0))
                 }
             ))
    
    optimal_weights = r.x
    pfo_mu = (optimal_weights.T @ mu).item((0, 0))
    pfo_risk = sqrt((optimal_weights.T @ covariance @ optimal_weights).item((0, 0)))
    
    return_value = 'Target:{}'.format(target_mu)
    return_value += ', Optimal Weights:{}'.format(optimal_weights)
    return_value += ', Mu:{:.4f}'.format(pfo_mu)
    return_value += ', SD:{:.4f}'.format(pfo_risk)
    return print(return_value)
```

```python
optimise_portfolio(0.045)
optimise_portfolio(0.045, corr_factor=1)
optimise_portfolio(0.045, corr_factor=1.25)
optimise_portfolio(0.045, corr_factor=1.5)
```

```python
def optimise_portfolio(risk_free_rate=0.005, corr_factor=1):

    mu = np.matrix([
        [0.02], 
        [0.07], 
        [0.15], 
        [0.20],
    ])

    sigma = np.diag([
        0.05, 
        0.12, 
        0.17, 
        0.25,
    ])

    correlation = np.matrix([
        [1.00, corr_factor * 0.30, corr_factor * 0.30, corr_factor * 0.30],
        [corr_factor * 0.30, 1.00, corr_factor * 0.60, corr_factor * 0.60],
        [corr_factor * 0.30, corr_factor * 0.60, 1.00, corr_factor * 0.60],
        [corr_factor * 0.30, corr_factor * 0.60, corr_factor * 0.60, 1.00],
    ])

    covariance = sigma @ correlation @ sigma

    start_weights = np.matrix([
        [0.25], 
        [0.25], 
        [0.25], 
        [0.25],
    ])

    def calc_sharpe(my_weights):
        pf_ret = (my_weights.T @ mu).item(0, 0)
        excess_return = pf_ret - risk_free_rate
        pf_sd = sqrt(my_weights.T @ covariance @ my_weights)
        sharpe = excess_return / pf_sd
        return -sharpe

    r = minimize(calc_sharpe, start_weights, tol=0.000000001,
             constraints=(
                 {
                     'type': 'eq', 
                     'fun': lambda weights: 1.0 - np.sum(weights)
                 },
             ))
    
    print('')
    print(r.message)
    optimal_weights = r.x
    pfo_mu = (optimal_weights.T @ mu).item((0, 0))
    pfo_excess_return = pfo_mu - risk_free_rate
    pfo_risk = sqrt((optimal_weights.T @ covariance @ optimal_weights).item((0, 0)))
    pfo_sharpe = pfo_excess_return / pfo_risk
    
    print('{:.4f} \t& {:.4f} \t& {:.4f} \t& {:.4f} \t& {:.4f} \t& {:.4f} \t& {:.4f} \t& {:.4f} \\\\'.format(risk_free_rate, optimal_weights[0], optimal_weights[1], 
                            optimal_weights[2], optimal_weights[3], pfo_mu, pfo_risk, pfo_sharpe))
    
    return_value = ' Risk Free:{}'.format(risk_free_rate)
    return_value += '\n Optimal Weights:{}'.format(optimal_weights)
    return_value += '\n Mu:{:.4f}'.format(pfo_mu)
    return_value += '\n SD:{:.4f}'.format(pfo_risk)
    return_value += '\n Sharpe: {:.4f}'.format(pfo_sharpe)
    return print(return_value)
```

```python
optimise_portfolio(0)
optimise_portfolio(0.0050)
optimise_portfolio(0.0075)
optimise_portfolio(0.0100)
optimise_portfolio(0.0150)
optimise_portfolio(0.0175)
```

Inverse optimisation problem where we generate weights randomly

```python
def question_2_part_2():
    n_weights = 1500
    
    mu = np.matrix([
        [0.02], 
        [0.07], 
        [0.15], 
        [0.20],
    ])
    
    sigma = np.diag([
        0.05, 
        0.12, 
        0.17, 
        0.25,
    ])
    
    # draw (time_steps x n_steps) random variables
    r = np.random.randn(4, n_weights)
    print('Sample weights: \n{}'.format(r[:,:5]))
    
    # find the sum of the four random weights
    s = np.sum(r, axis=0)
    print('Sample sum weights: \n{}'.format(s[:5]))
    
    # normalise each weight by dividing by sum of weights to ensure sum of weights add up to 1
    random_weights = (r.T / s[:, None]).T  # slice with None and combine with broadcasting
    print('Sample normalised weights: \n{}'.format(random_weights[:,:5]))
    
    # check random weights add up to 1
    # total number of random portfolios = 1500
    # ensure that range of weight values look reasonable
    check_random_weights = pd.DataFrame(random_weights.T)
    check_random_weights['sum'] = check_random_weights.sum(axis=1)
    print('Checking random weight statistics:\n{}'.format(check_random_weights.describe()))
    
    rand_pf_mu = random_weights.T @ mu
    print('Sample random portfolio expected returns:\n{}'.format(rand_pf_mu[:5]))
    
    correlation = np.matrix([
        [1.00, 0.30, 0.30, 0.30],
        [0.30, 1.00, 0.60, 0.60],
        [0.30, 0.60, 1.00, 0.60],
        [0.30, 0.60, 0.60, 1.00],
    ])

    covariance = sigma @ correlation @ sigma
    
    a = random_weights.T @ covariance @ random_weights
    rand_pf_sigma = np.sqrt(a.diagonal().T)
    print('Sample random portfolio volatility:\n{}'.format(rand_pf_sigma[:5]))
    
    # combine results into a dataframe so that we can more easily visualise
    df = pd.concat([
        pd.DataFrame(rand_pf_mu, columns=['pf_mu']), 
        pd.DataFrame(rand_pf_sigma, columns=['pf_sigma'])
    ], axis=1)
    # display the first five elements
    print(df.head())
    
    return df.plot(x=['pf_sigma'], y=['pf_mu'], kind='scatter', 
        xticks=np.arange(0, 0.25, .05), yticks=np.arange(0, 0.25, .05),
        alpha=0.2, figsize=(12, 8), title='Efficient Frontier generated by inverse optimisation'
       )
```

```python
question_2_part_2()
```

```python
corr_bump = -0.5

def plot_for_corr_bump(corr):
    
    n_weights = 1500
    
    mu = np.matrix([
        [0.02], 
        [0.07], 
        [0.15], 
        [0.20],
    ])
    sigma = np.diag([
        0.05, 
        0.12, 
        0.17, 
        0.25,
    ])
    correlation = np.matrix([
        [1.00, corr, corr, corr],
        [corr, 1.00, corr, corr],
        [corr, corr, 1.00, corr],
        [corr, corr, corr, 1.00],
    ])
    
    # draw (time_steps x n_steps) random variables
    r = np.random.randn(4, n_weights)
    print('Sample weights: \n{}'.format(r[:,:5]))
    
    # find the sum of the four random weights
    s = np.sum(r, axis=0)
    print('Sample sum weights: \n{}'.format(s[:5]))
    
    # normalise each weight by dividing by sum of weights to ensure sum of weights add up to 1
    random_weights = (r.T / s[:, None]).T  # slice with None and combine with broadcasting
    print('Sample normalised weights: \n{}'.format(random_weights[:,:5]))

    covariance = sigma @ correlation @ sigma
    rand_pf_mu = random_weights.T @ mu
    temp_df = random_weights.T @ covariance @ random_weights
    rand_pf_sigma = np.sqrt(temp_df.diagonal().T)
    df = pd.concat([pd.DataFrame(rand_pf_mu, columns=['pf_mu']), pd.DataFrame(rand_pf_sigma, columns=['pf_sigma'])], axis=1)
    return df.plot(x=['pf_sigma'], y=['pf_mu'], kind='scatter', 
            xticks=np.arange(0, 0.25, .05), yticks=np.arange(0, 0.25, .05),
            alpha=0.35, figsize=(12, 8), title='Average correlation = {}'.format(corr)
           )
```

```python
for x in [-0.5, -0.25, 0, 0.25, 0.5, 0.75, 1]:
    plot_for_corr_bump(x);
```
