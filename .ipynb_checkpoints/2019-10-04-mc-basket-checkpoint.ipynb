```python
%matplotlib inline
from IPython.display import display
import string
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_simulations(n_assets, n_time_steps, n_simulations, single_simulation_num):
    f, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 5))
    labels = [label for label in string.ascii_uppercase[:n_assets]]
    x_labels = index=np.linspace(0, T, n_time_steps + 1)
    df_single_sim = pd.DataFrame(spots[:, :, single_simulation_num].T, columns=labels, index=x_labels)  # set last element to the specific simulation to plot
    df_single_sim.plot(ax=ax1, title='Sample simulation number: {}'.format(single_simulation_num));
    df = pd.DataFrame(spots[:, :, :].mean(axis=2).T, columns=labels, index=x_labels)
    df.plot(ax=ax2, title='Expected paths of assets');
```

```python
def simulate_basket(initial_spots, mu, sigma, T, n_time_steps, n_sims, correlation_matrix):
    n_assets = len(initial_spots)
    dt = T / n_time_steps

    spots = np.zeros([n_assets, n_time_steps + 1, n_sims])  # Initial multi-dimensional spots array with zeros. 
    for asset_index, asset_initial_spot in enumerate(initial_spots):
        spots[asset_index, 0, :] = asset_initial_spot  # Initialise initial spot across all paths.

    sigma_diagonal_matrix = np.diag(sigma)
    covariance_matrix = sigma_diagonal_matrix @ correlation_matrix @ sigma_diagonal_matrix
    L = np.linalg.cholesky(covariance_matrix)

    print('T: {}, dt: {}, n time steps: {}, n assets: {}'.format(T, dt, n_time_steps, n_assets))
    labels = [label for label in string.ascii_uppercase[:n_assets]]
    display(pd.concat([
        pd.DataFrame(mu, columns=['r'], index=labels), 
        pd.DataFrame(sigma, columns=['vol'], index=labels), 
        pd.DataFrame([label for label in labels], columns=['correlation'], index=labels), 
        pd.DataFrame(correlation_matrix, columns=labels, index=labels)
    ], axis=1))

    # Simulate
    for time_step in range(int(n_time_steps)):
        Z = np.random.normal(0, 1, (n_assets, n_sims))
        X = L @ Z
        drift = (mu - 0.5 * sigma ** 2) * dt
        diffusion = np.sqrt(dt) * X
        growth_factor = np.exp(drift + diffusion.T).T
        spots[:, time_step + 1, :] = np.multiply(spots[:, time_step, :], growth_factor)
        
    return spots
```

```python
def payoff_basket_call(spots, K):
    F = spots[:, -1, :].mean(axis=0)
    return np.mean(np.maximum(F - 100, 0))
```

```python
n_assets = 4
initial_spots = [100] * n_assets
mu = np.array([0.05] * n_assets)
sigma = np.array([0.28] * n_assets)

T = 1
n_time_steps = 252

n_sims = 10000
corr = 0.7
correlation_matrix = np.full((n_assets, n_assets), corr)
np.fill_diagonal(correlation_matrix, 1)
# correlation_matrix = np.array([
#     [1.00, corr, corr, corr], 
#     [corr, 1.00, corr, corr], 
#     [corr, corr, 1.00, corr], 
#     [corr, corr, corr, 1.00]
# ])

spots = simulate_basket(initial_spots, mu, sigma, T, n_time_steps, n_sims, correlation_matrix)
plot_simulations(n_assets, n_time_steps, n_sims, 0)

print('basket call: {:0.4f}'.format(np.exp(-0.05) * payoff_basket_call(spots, 100)))

finals = spots[:, -1, :]
ranked_f = np.sort(finals, axis=0)
ranked_f_options = np.maximum(ranked_f-100, 0)
payoff_ranked_options = np.exp(-0.05) * np.mean(ranked_f_options, axis=1)
print('worst of call: {:0.4f}'.format(payoff_ranked_options[0]))
print('best of call: {:0.4f}'.format(payoff_ranked_options[-1]))
```
