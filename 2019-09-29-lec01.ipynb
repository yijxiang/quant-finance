```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linspace
from scipy.stats import norm

plt.rc('text', usetex=True)
plt.rc('font', family='serif', size=16)
```

```python
fig = plt.figure(figsize = (6.5,4))

x = linspace(-4,4)
y = norm.cdf(x)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.xlabel(r'$x$'); plt.ylabel(r'$\Phi(x)$')
```

```python
plt.subplot(1,2,2)
plt.plot(y,x)
plt.xlabel(r'$x$'); plt.ylabel(r'$\Phi^{-1}(x)$')

fig.tight_layout()
```

```python
fig = plt.figure(figsize = (6.5,3))
```

```python

x = linspace(-4,4)
y = norm.pdf(x)

plt.plot(x,y)
plt.xlabel(r'$x$'); plt.ylabel(r'$\phi(x)$')

for x0 in np.arange(0.25, 3.0, 0.5):
    y0 = norm.pdf(x0)
    plt.plot([ x0,  x0],    [0 , y0])
    plt.plot([-x0, -x0],    [0 , y0])
    plt.plot([ x0,  x0-0.5],[y0, y0])
    plt.plot([-x0, -x0+0.5],[y0, y0])

fig.tight_layout()
```
